# Preparar o dataset 

```python
import pandas as pd
```

```python
propertiesdata = pd.read_csv('datanprocess/molecules_properties.csv')
propertiesdata.columns
```

```python
propertiesdata['iscorante'] = propertiesdata['corante_00'] | propertiesdata['corante_max']
propertiesdata['iscorante'].value_counts()
propertiesdata.drop(columns=['corante_00', 'corante_max'])
```

```python
rdkit = pd.read_csv('datanprocess/rdkit_descriptors.csv')
rdkit.head()
```

```python
properties_rdkit = propertiesdata.merge(rdkit, on='cid', how='inner')
```
